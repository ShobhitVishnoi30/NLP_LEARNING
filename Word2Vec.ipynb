```python
import numpy as np
import pandas as pd
import gensim
import os
```

```python
from nltk import sent_tokenize
import nltk
from nltk.corpus import stopwords
from gensim.utils import simple_preprocess
nltk.download('stopwords')
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize

def remove_stopwords(sentence):
    # Get the list of English stopwords
    stop_words = set(stopwords.words('english'))

    # Tokenize the sentence
    word_tokens = word_tokenize(sentence)

    # Remove stopwords
    filtered_sentence = [word for word in word_tokens if word.lower() not in stop_words]

    # Join the words back into a sentence
    return ' '.join(filtered_sentence)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir('./data'):

    f = open(os.path.join('./data',filename))
    print(f)
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        sent=remove_stopwords(sent)
        story.append(simple_preprocess(sent))
```

```python
game_of_thrones_story = []
for filename in os.listdir('./gameOfThronesData'):
    print(os.path.join('./gameOfThronesData',filename))
    if os.path.join('./gameOfThronesData',filename)=="./gameOfThronesData/.ipynb_checkpoints":
      break
    f = open(os.path.join('./gameOfThronesData',filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        sent=remove_stopwords(sent)
        game_of_thrones_story.append(simple_preprocess(sent))
```

```python
model=gensim.models.Word2Vec(
    window=25,
    min_count=2
)
```

```python
model2=gensim.models.Word2Vec(
    window=25,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model2.build_vocab(game_of_thrones_story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model2.train(game_of_thrones_story, total_examples=model2.corpus_count, epochs=model2.epochs)
```
